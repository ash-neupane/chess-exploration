```python
def softmax(Z):
    """
    Z: (batch_size, seq_len, seq_len)
    """
    Z = Z - Z.max(axis=-1, keepdims=True)
    return np.exp(Z) / Z.sum(axis=-1, keepdims=True)

def self_attention(X, mask, W_KQV, Wout):
    """ 
    X: (batch_size, seq_len, d_model)
    mask: (batch_size, seq_len)
    W_KQV: (d_model, 3*d_model)
    Wout: (d_model, d_model)
    """
    K, Q, V = np.split(X @ W_KQV, 3, axis=1)
    attn = softmax(K @ Q.T / np.sqrt(X.shape[1]) + mask)
    return attn @ V @ Wout, attn 
```

```python
T, d = 100, 64
attn = torch.nn.MultiheadAttention(d, 1, bias=False, batch_first=True)
M = torch.triu(-float("inf")*torch.ones(T, T), 1)
X = torch.randn(1, T, d)
Y_, A_ = attn(X, X, X, attn_mask=M)
```
